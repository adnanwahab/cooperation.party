```python
import os


import json
from lxml import etree

def osm_to_json_houses(fp):
    # Parse the OSM XML
    tree = etree.parse(fp)
    root = tree.getroot()

    houses = []

    for elem in root:
        if elem.tag == "node" or elem.tag == "way":
            for tag in elem.findall('tag'):
                if tag.get("k") == "building" and tag.get("v") == "house":
                    house = {
                        "type": elem.tag,
                        "id": elem.get("id"),
                        "lat": elem.get("lat"),
                        "lon": elem.get("lon"),
                        "tags": {}
                    }

                    for tag in elem.findall('tag'):
                        house["tags"][tag.get("k")] = tag.get("v")

                    houses.append(house)

    return json.dump(houses, open(fp.replace('.osm', '.json'), 'w'), indent=4)

# Use the function
#json_houses = osm_to_json_houses("./osm_homes/Sydney.osm")
#print(json_houses)
lxml_this = glob.glob('./osm_homes/*_houses.osm')
lxml_this
for fp in lxml_this:
    osm_to_json_houses(fp)
```
